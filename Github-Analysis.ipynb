```python

# Group by product and sum the quantities
grouped_df = df.groupby('Product').sum()

# Plotting
fig, ax = plt.subplots(figsize=(8, 8))

# Plot a pie chart
ax.pie(grouped_df['Quantity'], labels=grouped_df.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral'])

# Adding a title
ax.set_title('Proportion of Shared Storage and Actions Usage')

# Equal aspect ratio ensures the pie chart is circular
ax.axis('equal')

# Show the plot
plt.show()
```
